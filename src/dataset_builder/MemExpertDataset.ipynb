```python
import os
import json
from PIL import Image
from datasets import Dataset
from datasets import load_dataset
```

```python
def gen(dir):
    dirj = os.path.join(dir, "json_data")
    diri = os.path.join(dir, "images")

    for file in os.listdir(dirj):
        with open(os.path.join(dirj, file), 'r', encoding='utf-8') as f:
            data = json.load(f)

        yield {
            "image": Image.open(os.path.join(diri, f"{file.split('.')[0]}.jpg")),
            "description": data["description"],
            "alt": data["image:alt"]
        }
```

```python
ds = Dataset.from_generator(gen, gen_kwargs={"dir": "."})
```
